# Day 2 — Python for data + getting your dataset

Project: Predict Disney+ content rating (TV-G, PG, TV-14, ...) from genre, runtime, release year, and country.

This notebook is today's pandas practice only: load, filter, groupby. Day 3's cleaning pipeline lives separately in `clean_data.py` / `clean_data.ipynb`, since that's a reusable artifact later stages import — not exploratory scratch work.

```python
import pandas as pd
from pathlib import Path

# Raw data lives at the project root, one level up from stage01/
# (confirmed from the actual repo layout -- adjust here if that changes).
SCRIPT_DIR = Path.cwd()
RAW_PATH = SCRIPT_DIR.parent / "disney_plus_titles.csv"

df = pd.read_csv(RAW_PATH)
print(f"Shape: {df.shape[0]} rows, {df.shape[1]} columns")
df.dtypes
```

## Filter examples

```python
# How many titles came out in the last 5 years of the catalog?
recent = df[df["release_year"] >= 2017]
print(f"Titles released 2017 or later: {len(recent)} of {len(df)}")

# Isolate just the TV shows
tv_shows = df[df["type"] == "TV Show"]
print(f"TV Shows: {len(tv_shows)} | Movies: {len(df) - len(tv_shows)}")
```

## Groupby examples

```python
# How many titles per rating category?
by_rating = df.groupby("rating", dropna=False)["title"].count().sort_values(ascending=False)
print(f"Title count by rating:\n{by_rating}")
```

```python
# Average release year per content type -- do TV shows skew newer?
avg_year_by_type = df.groupby("type")["release_year"].mean()
avg_year_by_type
```
